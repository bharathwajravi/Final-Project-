```python
from transformers import pipeline

# Load the summarization model
summarizer = pipeline("summarization", model="t5-small")

# Test the summarizer
text = "Your long text here."
summary = summarizer(text, max_length=150, min_length=30, do_sample=False)
print(summary[0]['summary_text'])
```

```python
from transformers import pipeline

# Load the text generation model
generator = pipeline("text-generation", model="gpt2") # Use gpt2 instead of gpt-2

# Test the predictor
text = "Once upon a time"
prediction = generator(text, max_length=50, num_return_sequences=1)
print(prediction[0]['generated_text'])
```

```python
from transformers import pipeline

# Load the conversational model
# Use 'conversational' for newer versions or specify a model
chatbot = pipeline("text-generation", model="facebook/blenderbot-400M-distill")

# Test the chatbot
text = "Hello, how are you?" # Assign the text directly to the text variable
response = chatbot(text) # Pass the text to the chatbot
print(response[0]['generated_text']) # Access the generated text from the response
```

```python
from transformers import pipeline

# Load the sentiment analysis model
sentiment_analyzer = pipeline("sentiment-analysis")

# Test the sentiment analyzer
text = "I love this product!"
sentiment = sentiment_analyzer(text)
print(sentiment[0])
```

```python
from transformers import pipeline

# Load the question answering model
qa_model = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")

# Test the QA model
context = "Google Colab is a cloud service from Google."
question = "What is Google Colab?"
result = qa_model(question=question, context=context)
print(result['answer'])
```
